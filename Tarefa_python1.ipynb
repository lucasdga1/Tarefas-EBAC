```python
# Usuário escolhe o primeiro número
print("Olá! Vamos executar alguns cálculos\n")
num1 = float(input("Escolha seu primeiro número: "))
# Usuário escolhe a operação
operacao = str(input("\nEscolha a operação que você deseja realizar entre adição (A), subtração (S), multiplicação (M) e divisão (D): "))
# Usuário escolhe o segundo número
num2 = float(input("\nEscolha seu segundo número: "))

# Execução da operação
if operacao.capitalize() == "A":
 result = num1 + num2
 print("\nAqui está seu resultado: ", result)
elif operacao.capitalize() == "S":
  result = num1 - num2
  print("\nAqui está seu resultado: ", result)
elif operacao.capitalize() == "M":
  result = num1 * num2
  print("\nAqui está seu resultado: ", result)
elif operacao.capitalize() == "D":
  result = num1 / num2
  print("\nAqui está seu resultado: ", result)
else:
  print("\nPor favor, escolha uma operação dentre as opções apresentadas")
```
